## Data cleaning

```python
#Extracting dates from ListingCreationDate column
df['year'] = df.ListingCreationDate.apply(lambda x: x.split("-")[0]).astype(str)
df['month'] = df.ListingCreationDate.apply(lambda x: x.split("-")[1]).astype(str)
```

```python
#Creating a subset dataframe for the features we want to focus on
df2 = df[['ListingNumber','year','month','Term','Occupation', 'LoanStatus', 'ListingCategory (numeric)' , 'ProsperRating (Alpha)' , 'EmploymentStatus' , 'EmploymentStatusDuration', 'IncomeRange' ,'StatedMonthlyIncome' ,'MonthlyLoanPayment' ]]



```

```python
df2.sample(5)
```

```python
#Renaming cols
df2 = df2.rename(columns = {'ListingCategory (numeric)': 'ListingCategory'})
df2 = df2.rename(columns = {'ProsperRating (Alpha)': 'ProsperRating'})
```
